```python
# Step 1: Install necessary tools
!apt-get install poppler-utils -y
!apt install tesseract-ocr -y
!pip install pdf2image pytesseract pillow

# Step 2: Import required libraries
from pdf2image import convert_from_path
from PIL import Image
import pytesseract
import os

# Step 3: Upload your PDF
from google.colab import files
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# Step 4: Convert PDF to images
images = convert_from_path(pdf_path)

# Step 5: Extract text from each image
os.makedirs("pdf_pages", exist_ok=True)
texts = []

for i, image in enumerate(images):
    image_path = f"pdf_pages/page_{i+1}.png"
    image.save(image_path, "PNG")

    # OCR: extract text from image
    text = pytesseract.image_to_string(Image.open(image_path))
    texts.append((image_path, text))

# Step 6: Display extracted text
for path, content in texts:
    print(f"\n--- Text from {path} ---\n{content}\n")
```
